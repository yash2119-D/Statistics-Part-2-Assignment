### 1. What is hypothesis testing in statistics?

Hypothesis testing is a formal statistical method used to evaluate assumptions or claims about a population based on sample data. It is a decision-making process that determines whether to accept or reject a stated hypothesis. The process begins by formulating two competing hypotheses: the null hypothesis (“H₀”), which assumes no effect or no difference, and the alternative hypothesis (“Hₐ”), which represents a significant effect or difference. Data from a sample is then analyzed, and the results are used to infer conclusions about the population as a whole.

Hypothesis testing is fundamental in many disciplines, such as medicine, social sciences, and business, where decisions must be made based on limited data. For example, a pharmaceutical company might test whether a new drug is more effective than an existing one, or a business might analyze whether a new marketing strategy improves sales. By following a systematic approach, hypothesis testing minimizes bias and provides a framework for making data-driven decisions.

---

### 2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

The **null hypothesis (H₀)** is the default assumption that there is no effect, no difference, or no relationship between the variables being studied. It represents the status quo or a baseline measurement that serves as the point of comparison. For example, if we are testing whether a new teaching method improves students' scores, the null hypothesis might state that the mean scores of students using the new method are equal to those using the traditional method.

The **alternative hypothesis (Hₐ)** is the statement that contradicts the null hypothesis. It suggests that there is an effect, a difference, or a relationship between the variables. Using the same example, the alternative hypothesis would state that the mean scores of students using the new method are higher than those using the traditional method.

Key differences:
- The null hypothesis is generally assumed true until evidence suggests otherwise, whereas the alternative hypothesis is what researchers aim to support.
- The null hypothesis is tested directly, and the evidence is evaluated to determine whether it should be rejected in favor of the alternative hypothesis.

---

### 3. What is the significance level in hypothesis testing, and why is it important?

The **significance level (α)** is a threshold used in hypothesis testing to determine whether the null hypothesis should be rejected. It represents the probability of making a Type 1 error—rejecting the null hypothesis when it is actually true. Commonly used significance levels are 0.05 (5%), 0.01 (1%), or 0.10 (10%).

The significance level is important because it defines the criteria for decision-making. A lower significance level indicates stricter criteria for rejecting the null hypothesis, reducing the likelihood of a Type 1 error but increasing the risk of a Type 2 error (failing to reject a false null hypothesis). For example, in clinical trials, a lower significance level might be chosen to minimize the risk of falsely approving an ineffective treatment.

---

### 4. What does a P-value represent in hypothesis testing?

The **P-value** is a measure of the strength of evidence against the null hypothesis. It represents the probability of observing results as extreme as, or more extreme than, the sample data, assuming the null hypothesis is true. A smaller P-value indicates stronger evidence against the null hypothesis, suggesting that the observed data is unlikely to occur by random chance.

The P-value is central to hypothesis testing because it quantifies the compatibility of the data with the null hypothesis. For example, if a study tests whether a new drug reduces blood pressure and yields a P-value of 0.03, it suggests that there is only a 3% chance of observing the data if the drug had no effect.

---

### 5. How do you interpret the P-value in hypothesis testing?

Interpreting the P-value involves comparing it to the significance level (α):
- **If the P-value ≤ α**: Reject the null hypothesis. This indicates that the observed results are statistically significant and unlikely to have occurred by random chance.
- **If the P-value > α**: Fail to reject the null hypothesis. This suggests that there is insufficient evidence to conclude that the null hypothesis is false.

For instance, a P-value of 0.02 in a test with a significance level of 0.05 would lead to rejecting the null hypothesis. However, the same P-value in a test with a significance level of 0.01 would not provide enough evidence to reject the null hypothesis.

---

### 6. What are Type 1 and Type 2 errors in hypothesis testing?

- **Type 1 Error (α)**: Occurs when the null hypothesis is rejected even though it is true. This is a false positive, meaning a researcher concludes that an effect exists when it does not. For example, concluding that a new drug is effective when it is not.

- **Type 2 Error (β)**: Occurs when the null hypothesis is not rejected even though it is false. This is a false negative, meaning a researcher fails to detect an actual effect. For example, concluding that a drug has no effect when it actually does.

Balancing these errors is a critical aspect of experimental design. Decreasing the risk of one type of error often increases the risk of the other.

---

### 7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

- **One-tailed test**: Examines whether a parameter is greater than or less than a certain value in a specific direction. For example, testing whether a new teaching method improves scores (greater than).

- **Two-tailed test**: Examines whether a parameter is different from a certain value, regardless of direction. For example, testing whether a new teaching method affects scores, either positively or negatively.

A one-tailed test is more sensitive to detecting effects in one direction but ignores effects in the opposite direction. A two-tailed test is more conservative, as it accounts for deviations in both directions.

---

### 8. What is the Z-test, and when is it used in hypothesis testing?

The **Z-test** is a statistical test used to determine whether there is a significant difference between the sample mean and the population mean when the population variance is known. It is appropriate for large sample sizes (≥30) and assumes that the data follows a normal distribution.

The Z-test is commonly used in:
- Testing hypotheses about population means or proportions.
- Comparing a sample mean to a known population mean.

---

### 9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

The **Z-score** represents the number of standard deviations a data point is from the population mean. In hypothesis testing, it is used to determine how extreme the observed data is under the null hypothesis.

A Z-score allows researchers to assess whether the sample statistic falls within the range of expected values under the null hypothesis. Larger absolute values of the Z-score indicate that the data is farther from the mean and more likely to provide evidence against the null hypothesis.

---

### 10. What is the T-distribution, and when should it be used instead of the normal distribution?

The **T-distribution** is a probability distribution similar to the normal distribution but with heavier tails, meaning it is more prone to extreme values. It is used when:
- The sample size is small (≤30).
- The population variance is unknown.

The T-distribution accounts for the uncertainty in estimating the population standard deviation from the sample data, making it more appropriate for small sample sizes.

---

### 11. What is the difference between a Z-test and a T-test?

- **Z-test**: Used when the population variance is known and the sample size is large.
- **T-test**: Used when the population variance is unknown and/or the sample size is small.

---

### 12. What is the T-test, and how is it used in hypothesis testing?

The **T-test** is a statistical test used to compare the means of one or more groups to determine whether they are significantly different. Types of T-tests include:
- One-sample T-test: Compares the sample mean to a known population mean.
- Independent two-sample T-test: Compares the means of two independent groups.
- Paired T-test: Compares the means of two related groups.

---

### 13. What is the relationship between Z-test and T-test in hypothesis testing?

The Z-test and T-test are related as both evaluate the difference between observed data and expected values under the null hypothesis. However, they differ in their assumptions about variance and sample size. As sample size increases, the T-distribution approaches the normal distribution, making the tests yield similar results.

---

### 14. What is a confidence interval, and how is it used to interpret statistical results?

A **confidence interval (CI)** is a range of values within which the true population parameter is likely to fall, given a specific level of confidence (e.g., 95%). It provides a measure of the precision of an estimate and helps assess the uncertainty in data analysis.

---

### 15. What is the margin of error, and how does it affect the confidence interval?

The **margin of error** represents the maximum amount by which the sample estimate may differ from the true population parameter. It affects the width of the confidence interval: a larger margin of error results in a wider interval, indicating greater uncertainty.

---

### 16. How is Bayes' Theorem used in statistics, and what is its significance?

Bayes' Theorem calculates the probability of an event based on prior knowledge of related conditions. It is significant in updating probabilities as new evidence is observed. Applications include machine learning, medical diagnosis, and decision-making.

---

### 17. What is the Chi-square distribution, and when is it used?

The **Chi-square distribution** is used to analyze categorical data and test hypotheses about relationships between variables. It is applied in tests of independence, goodness of fit, and homogeneity.

---

### 18. What is the Chi-square goodness of fit test, and how is it applied?

The **Chi-square goodness of fit test** evaluates whether an observed frequency distribution matches an expected distribution. It assesses whether a theoretical model fits the data.

---

### 19. What is the F-distribution, and when is it used in hypothesis testing?

The **F-distribution** is used to compare variances between groups. It is applied in tests like ANOVA and regression analysis to determine if group differences are significant.

---

### 20. What is an ANOVA test, and what are its assumptions?

**ANOVA (Analysis of Variance)** tests differences in means across multiple groups. Assumptions include normality, homogeneity of variances, and independence of observations.

---

### 21. What are the different types of ANOVA tests?

- **One-way ANOVA**: Tests one factor with multiple levels.
- **Two-way ANOVA**: Tests two factors and their interaction.
- **Repeated measures ANOVA**: Tests within-subject variations.

---

### 22. What is the F-test, and how does it relate to hypothesis testing?

The **F-test** evaluates the ratio of variances between groups to determine if the differences are statistically significant. It is central to ANOVA and regression analysis.

```python
#Q :- 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
#Answer :-
import scipy.stats as stats

# Data
population_mean = 100
population_std = 15
sample_mean = 105
sample_size = 30

# Calculate Z-score
z_score = (sample_mean - population_mean) / (population_std / (sample_size ** 0.5))

# Calculate the p-value for a two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

```python
#Q :- 3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.
#Answer :-
def one_sample_z_test(sample, population_mean, population_std):
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    return z_score, p_value

# Simulating sample data
np.random.seed(42)
sample_data = np.random.normal(100, 15, 30)

# Perform the Z-test
z, p = one_sample_z_test(sample_data, 100, 15)
print(f"Z-score: {z}")
print(f"P-value: {p}")
```

```python
#Q :- 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
#Answer :-
def visualize_errors(true_mean, sample_mean, sample_std, sample_size, alpha=0.05):
    z_critical = stats.norm.ppf(1 - alpha / 2)
    x = np.linspace(true_mean - 4 * sample_std, true_mean + 4 * sample_std, 1000)
    y = stats.norm.pdf(x, true_mean, sample_std / np.sqrt(sample_size))

    # Rejection region
    rejection_low = true_mean - z_critical * (sample_std / np.sqrt(sample_size))
    rejection_high = true_mean + z_critical * (sample_std / np.sqrt(sample_size))

    plt.plot(x, y, label="True Distribution")
    plt.axvline(rejection_low, color='red', linestyle='--', label="Rejection Boundary")
    plt.axvline(rejection_high, color='red', linestyle='--')

    plt.fill_between(x, y, where=(x < rejection_low) | (x > rejection_high), color='red', alpha=0.3, label="Type I Error")
    plt.title("Type I and Type II Errors")
    plt.legend()
    plt.show()

# Example usage
visualize_errors(100, 105, 15, 30)
```

```python
#Q :- 6. Write a Python program to perform an independent T-test and interpret the results.
#Answer :-
from scipy.stats import ttest_ind

# Simulating two independent samples
np.random.seed(42)
sample1 = np.random.normal(100, 15, 30)
sample2 = np.random.normal(105, 15, 30)

# Perform independent T-test
t_stat, p_value = ttest_ind(sample1, sample2)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: The means of the two samples are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the sample means.")
```

```python
#Q :- 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
#Answer :-
def confidence_interval(sample, confidence=0.95):
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    sample_size = len(sample)
    critical_value = stats.t.ppf((1 + confidence) / 2, df=sample_size - 1)
    margin_of_error = critical_value * (sample_std / np.sqrt(sample_size))
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    return lower_bound, upper_bound

# Simulate sample data
np.random.seed(42)
sample_data = np.random.normal(100, 15, 30)

# Calculate confidence interval
ci = confidence_interval(sample_data)
print(f"95% Confidence Interval: {ci}")

# Significance
print("The confidence interval provides a range within which the true population mean is likely to fall with 95% confidence.")
```

```python
#Q :- 10. Write a Python program to calculate the margin of error for a given confidence level using sample data.
#Answer :-
import scipy.stats as stats
import math

# Inputs
confidence_level = 0.95
sample_std_dev = 15  # Sample standard deviation
sample_size = 50

# Calculate Z-score for the given confidence level
z_score = stats.norm.ppf((1 + confidence_level) / 2)

# Calculate standard error
standard_error = sample_std_dev / math.sqrt(sample_size)

# Margin of error
margin_of_error = z_score * standard_error

print(f"Margin of Error: {margin_of_error:.2f}")
```

```python
#Q :- 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
#Answer :-
# Define probabilities
prior = 0.6  # P(A)
likelihood = 0.8  # P(B|A)
evidence = 0.5  # P(B)

# Calculate posterior using Bayes' theorem: P(A|B) = (P(B|A) * P(A)) / P(B)
posterior = (likelihood * prior) / evidence

print(f"Posterior Probability: {posterior:.2f}")
#Explanation :-
#Bayes' Theorem relates the conditional probability of an event 𝐴 given 𝐵, using prior knowledge of P(A), 𝑃(𝐵∣𝐴), and P(B).
```

```python
#Q :- 12. Perform a Chi-square test for independence between two categorical variables in Python.
#Answer :-
import scipy.stats as stats
import numpy as np

# Observed frequencies
observed = np.array([[50, 30], [20, 100]])

# Perform Chi-square test for independence
chi2, p, dof, expected = stats.chi2_contingency(observed)

print(f"Chi-square Statistic: {chi2:.2f}")
print(f"P-value: {p:.2e}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies: \n{expected}")
```

```python
#Q :- 13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.
#Answer :-
# Observed frequencies
observed = np.array([[50, 30], [20, 100]])

# Marginal totals
row_totals = observed.sum(axis=1)
col_totals = observed.sum(axis=0)
grand_total = observed.sum()

# Calculate expected frequencies
expected = np.outer(row_totals, col_totals) / grand_total

print(f"Expected Frequencies: \n{expected}")
```

```python
#Q :- 14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
#Answer :-
# Observed and expected data
observed = np.array([60, 54, 46, 41])
expected = np.array([50, 50, 50, 50])

# Adjust expected frequencies to match the total observed frequency
expected = expected * sum(observed) / sum(expected) # This line adjusts the expected frequencies
                                                 # to ensure their sum equals the sum of observed frequencies.

# Perform Chi-square goodness-of-fit test
chi2, p = stats.chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square Statistic: {chi2:.2f}")
print(f"P-value: {p:.2e}")
```

```python
#Q :- 16. Implement an F-test using Python to compare the variances of two random samples.
#Answer :-
# Generate two samples
sample1 = np.random.normal(50, 10, 30)
sample2 = np.random.normal(52, 15, 30)

# Calculate variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Perform F-test
f_statistic = var1 / var2
df1, df2 = len(sample1) - 1, len(sample2) - 1
p_value = stats.f.cdf(f_statistic, df1, df2)

print(f"F-statistic: {f_statistic:.2f}")
print(f"P-value: {p_value:.2e}")
```

```python
#Q :- 19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.
#Answer :-
from scipy.stats import shapiro, levene

def check_anova_assumptions(*groups):
    # Check normality using Shapiro-Wilk test
    for i, group in enumerate(groups, 1):
        stat, p = shapiro(group)
        print(f"Group {i} Normality Test: Statistic={stat:.2f}, P-value={p:.2e} {'(Pass)' if p > 0.05 else '(Fail)'}")

    # Check homogeneity of variances using Levene's test
    stat, p = levene(*groups)
    print(f"Levene's Test for Equal Variances: Statistic={stat:.2f}, P-value={p:.2e} {'(Pass)' if p > 0.05 else '(Fail)'}")

# Sample data
group1 = [4, 5, 6, 7, 8]
group2 = [5, 6, 7, 8, 9]
group3 = [6, 7, 8, 9, 10]

check_anova_assumptions(group1, group2, group3)
```

```python
#Q :- 20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.
#Answer :-
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns

# Create dataset
data = {
    'Factor_A': ['Low', 'Low', 'Low', 'High', 'High', 'High'] * 2,
    'Factor_B': ['Male', 'Male', 'Female', 'Male', 'Male', 'Female'] * 2,
    'Scores': [85, 90, 88, 70, 75, 80, 87, 92, 89, 72, 78, 81]
}
df = pd.DataFrame(data)

# Perform two-way ANOVA
model = ols('Scores ~ C(Factor_A) + C(Factor_B) + C(Factor_A):C(Factor_B)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)

# Visualization
sns.boxplot(x='Factor_A', y='Scores', hue='Factor_B', data=df)
plt.title('Two-way ANOVA Results')
plt.show()
```
